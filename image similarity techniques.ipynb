```python
import numpy as np
Xn=np.reshape(X,(10000,784))
```

```python
from sklearn.metrics.pairwise import cosine_similarity as cs
import matplotlib.pyplot as plt
import pandas as pd
```

```python
cossim=cs(Xn)
```

```python
cosdataframe=pd.DataFrame(cossim)
```

```python
prod=cosdataframe.iloc[543]
```

```python
imshow(X[543])
```

```python
#argsort returns the index positions of the values being sorted if they were to be sorted
similar_images=np.argsort(-prod)[0:10]
```

```python
similar_images
```

```python
plt.figure(1,figsize=(20,100))
for i in range(0,10):
    plt.subplot(1,10,i+1)
    plt.imshow(X[similar_images[i]])
plt.show()    
```
